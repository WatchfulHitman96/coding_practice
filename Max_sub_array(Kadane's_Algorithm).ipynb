```python
def max_sub_array(nums):
    # Initialize variables with the first element
    max_sum = nums[0]
    current_sum = nums[0]

    # Track indices for the result
    start_index = 0
    end_index = 0

    # Track the start of the current running calculation
    temp_start_index = 0

    # Loop from the second element to the end of the list
    for i in range(1, len(nums)):

        # Logic: Should we restart the subarray at current number nums[i],
        # or add nums[i] to the existing sum?
        if nums[i] > current_sum + nums[i]:
            current_sum = nums[i]
            temp_start_index = i  # Reset start position
        else:
            current_sum += nums[i]

        # Logic: Update the global maximum if current is higher
        if current_sum > max_sum:
            max_sum = current_sum
            start_index = temp_start_index
            end_index = i

    # Return the sum and the actual list slice
    # end_index + 1 is used because Python slices are exclusive at the end
    return max_sum, nums[start_index : end_index + 1]

# --- Example Usage ---

# Example 1
nums1 = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
result_sum, result_arr = max_sub_array(nums1)
print("Sum:", result_sum)
print("Array:", result_arr)

# Example 2
nums2 = [5, 4, -1, 7, 8]
result_sum, result_arr = max_sub_array(nums2)
print("Sum:", result_sum)
print("Array:", result_arr)
```
